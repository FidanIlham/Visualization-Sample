```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('/Users/FidanIlham/Downloads/Orders_File_Alaa_Saved3.csv')
```

```python
country_count  = sales.groupby('Country1').sum()['Sales'].sort_values(ascending=False)
country_count = country_count[:20,]
plt.figure(figsize=(10,5))
sns.barplot(country_count.index, country_count.values, alpha=0.8)
plt.title('Sales in top 20 countries in the World')
plt.ylabel('\$\$Sales in USD I think....', fontsize=12)
plt.xticks(rotation = 90)
plt.xlabel('Country', fontsize=12)
plt.show()
```
